```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

```

```python
df = pd.read_csv("C:/Users/harshitha/Downloads/Task 2 AIML/Titanic-Dataset.csv")
df.head()

```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.hist(bins=20, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
numeric_cols = ['Age', 'Fare']
for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Pclass']], hue='Survived', palette='Set2')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Survived', data=df, palette='coolwarm', legend=False)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='viridis')
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='plasma')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
fig = px.scatter(df, x='Age', y='Fare', color='Survived',
                 title='Age vs Fare (colored by Survival)')
fig.show()
```
